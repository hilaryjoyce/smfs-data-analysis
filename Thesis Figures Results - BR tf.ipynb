```python
def plot_lcd_tf_subplots(sub_list, coa, max_x, lcd_y=0.15, tf_y = 300, min_lc= 0):
    from numpy import ceil
    import matplotlib.pyplot as plt
    cpool = ["#1F78B4", "#E31A1C"]
    lcd_list = v.cluster_lcd_list(sub_list, coa)
    tf_list = v.cluster_tf_list(sub_list, coa)
    cols = 3
    rows = ceil(len(lcd_list)/3.0)
    height = rows * 1.8
    width = cols * 4
    i = 1
    plt.figsize(10, height)
    while i <= len(lcd_list):
        lcd = lcd_list[i-1];
        tf = tf_list[i-1];
        subplot(rows, cols, i);
        plot(lcd[0], lcd[1], color=cpool[1], lw=1);
        ylim(min_lc, lcd_y);
        text(max_x*0.9, lcd_y*0.85, '%d' % i, va='top', ha='right')
        plot(tf[0], tf[1], color=cpool[0], alpha=0.6)
        ylim(-100, tf_y)
        xlim(0, max_x);
        if (i-1)%3 == 0:
            ylabel('Force (pN)')
        else:
            yticks([-100, -50, 0, 50, 100, 150], ['','','','','',''])
        if i > 9:
            xlabel('Separation (nm)')
            xticks([0, 50, 100, 150, 200])
        else:
            xticks([0, 50, 100, 150, 200], ['','','','',''])
        i = i+1;
        
    #plt.tight_layout(pad=2);
    #plt.subplots_adjust(top=0.96);
    return plt.gcf();
```

```python
lcd_folder = '../Data/bacteriorhodopsin/Transform_analysis/Parameters_p0.4_minTssVar_minForce45/Density_text_max600/'
lcd_files = glob(lcd_folder+'*.txt')
lcd_list = []
for cluster in good_signal_indexes:
    lcd_list.append([c.load2Col(lcd_files[x]) for x in cluster])
```

```python
peaks_list = []
for cluster in lcd_list:
    peaks = []
    for curve in cluster:
        maxima_x, maxima_y = findPeaks(curve[0], curve[1], 0.25)
        peaks.append([maxima_x, maxima_y])
    peaks_list.append(peaks)
```

```python
shifted_peaks_list = []
for [curves, shifts] in zip(peaks_list, shift_list):
    shifted_peaks = []
    for (peaks, s) in zip(curves, shifts):
        shift_x = [p+s for p in peaks[0]]
        shifted_peaks.append([shift_x, peaks[1]])
    shifted_peaks_list.append(shifted_peaks)
```

```python
all_shift_peaks_list = []
for peaks in shifted_peaks_list:
    for p in peaks:
        all_shift_peaks_list = all_shift_peaks_list + p[0]

all_peaks_list = []
for peaks in peaks_list:
    for p in peaks:
        all_peaks_list = all_peaks_list + p[0]
```

```python
last_shift_peak_list = []
for peaks in shifted_peaks_list:
    last_shift_peak_list.append([max(peak[0]) for peak in peaks])

last_peak_list = []
for peaks in peaks_list:
    last_peak_list.append([max(peak[0]) for peak in peaks])
```
