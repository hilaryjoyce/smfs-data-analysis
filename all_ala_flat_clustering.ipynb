```python
def cluster_tf_list(curve_list, coa):
    tf_files = [coa.tss_force_files[x] for x in curve_list]
    tf_list = []
    for file in tf_files:
        tf_list.append(c.load2Col(file))
    return tf_list

def cluster_lcd_list(curve_list, coa):
    coa_files = coa.list_density_files()
    lcd_files = [coa_files[x] for x in curve_list]
    lcd_list = []
    for file in lcd_files:
        lcd_list.append(c.load2Col(file))
    return lcd_list

def plot_subplots(data_list, id_list, max_x):
    cols = 4
    rows = ceil(len(data_list)/4.0)
    height = rows * 2.2
    width = cols * 3
    i = 1
    figsize(width, height)
    while i <= len(data_list):
        data = data_list[i-1];
        subplot(rows, cols, i);
        plot(data[0], data[1]);
        xlim(0, max_x);
        i = i+1;
    tight_layout(pad=2);
    subplots_adjust(top=0.96);
    return gcf();

def plot_tf_subplots(sub_list, coa, max_x=100, max_y=300):
    data_list = cluster_tf_list(sub_list, coa)
    cols = 4
    rows = ceil(len(data_list)/4.0)
    height = rows * 2.2
    width = cols * 3
    i = 1
    figsize(width, height)
    while i <= len(data_list):
        data = data_list[i-1];
        subplot(rows, cols, i);
        plot(data[0], data[1]);
        xlim(0, max_x);
        ylim(-200,max_y)
        text(max_x/1.2, max_y/1.2, '%d' % i)
        i = i+1;
    tight_layout(pad=2);
    subplots_adjust(top=0.96);
    return gcf();

def plot_lcd_subplots(data_list, coa, max_x=100, max_y=0.2):
    data_list = cluster_lcd_list(data_list, coa)
    cols = 4
    rows = ceil(len(data_list)/4.0)
    height = rows * 2.2
    width = cols * 3
    i = 1
    figsize(width, height)
    while i <= len(data_list):
        data = data_list[i-1];
        subplot(rows, cols, i);
        plot(data[0], data[1]);
        xlim(0, max_x);
        ylim(0, max_y);
        text(max_x/1.2, max_y/1.2, '%d' % i)
        i = i+1;
    tight_layout(pad=2);
    subplots_adjust(top=0.96);
    return gcf();
```

```python
cpool = ["#1F78B4", "#E31A1C", "#A6CEE3", "#FB9A99", "#7BCCC4", "#B2DF8A", "#33A02C", "#02818A", "#FF7F00", "#FDBF6F", "#CAB2D6", "#6A3D9A", "#BFD3E6", "#8C96C6"]
def plot_lcd_tf_subplots(sub_list, coa, max_x, lcd_y=0.15, tf_y = 300):
    lcd_list = cluster_lcd_list(sub_list, coa)
    tf_list = cluster_tf_list(sub_list, coa)
    cols = 4
    rows = ceil(len(lcd_list)/4.0)
    height = rows * 2.2
    width = cols * 3
    i = 1
    figsize(width, height)
    while i <= len(lcd_list):
        lcd = lcd_list[i-1];
        tf = tf_list[i-1];
        lc_ax = subplot(rows, cols, i);
        lc_ax.plot(lcd[0], lcd[1], color=cpool[1], lw=1);
        lc_ax.set_ylim(0, lcd_y);
        lc_ax.text(max_x/1.2, lcd_y/1.2, '%d' % i)
        tf_ax = lc_ax.twinx()
        tf_ax.plot(tf[0], tf[1], color=cpool[0], alpha=0.6)
        tf_ax.set_ylim(-100, tf_y)
        lc_ax.set_xlim(0, max_x);
        i = i+1;
    tight_layout(pad=2);
    subplots_adjust(top=0.96);
    return gcf();
```

## Working area for getting average cluster density

```python
lc_d= s33_4.get_Lc_density_arrays('left')
ishifts = s33_4.list_initial_shifts('left')
density = [x[1] for x in lc_d]
```

```python
def cluster_Lc_shifted(ishifts, Lc_density_list):
    shift_list = ishifts
    Lc_list = [x[0] for x in Lc_density_list]
    initial_shifts = ishifts
    return [[Lc+shift for Lc in Lc_array] for [Lc_array, shift] in zip(Lc_list, initial_shifts)]
```

```python
def shift_to_index(shift, grid=0.25):
    return int(shift/grid)
```

```python
shift_Lc = cluster_Lc_shifted(ishifts, lc_d)
```

```python
def average_cluster_density(ishifts, Lc_density_list):
    ''' Takes a cluster number and a set of flat clusters, the shift details and the 
    density list, and returns the average_density '''
    from numpy import copy
    shift_list = ishifts
    density_list = [x[1] for x in Lc_density_list]
    N = len(density_list)
    M = len(density_list[0])
    average_density = copy(density_list[0])
    k = 1
    while k < N:
        ind_shift = shift_to_index(shift_list[k-1])
        if ind_shift <= 0:
            j = 0
            while j < M+ind_shift:
                average_density[j] = average_density[j] + density_list[k][j-ind_shift]
                j = j+1
        else:
            j = ind_shift
            while j < M:
                average_density[j] = average_density[j] + density_list[k][j-ind_shift]
                j = j+1
        k = k+1
    average_density = average_density/N
    return average_density
```

```python
av_d = average_cluster_density(ishifts, lc_d)
```
