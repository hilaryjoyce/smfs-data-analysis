## Keeping track of curved vs. flat

For the combined flat and curved sample files:
    
- no letter, a, b, c, and d are flat files
- e, f, g, and h are curved files

```python
print len(coa.list_curve_names())
name_list = coa.list_curve_names()
```

```python
is_curved = zeros(len(name_list))
is_flat = zeros(len(name_list))
curved_letters = asarray(['e', 'f', 'g', 'h'])
flat_letters = asarray(['a', 'b', 'c', 'd'])
```

```python
i = 0
for curve in name_list:
    if any(curve[0] == curved_letters):
        is_curved[i] = 1
    else:
        is_flat[i] = 1
    i = i+1
```

- is_curved is an array of 0 or 1
- is_flat is an array of 0 or 1

## Looking at clusters

```python
co_cut = 0.65
F = c.FlatClusters(coa, co_cut)
```

```python
print F
F.print_clusters_by_size(20)
```

```python
cluster = c.Cluster(F, 189)
print cluster
```

```python
cluster_indexes = F.curves_by_cluster()
cluster_sizes = F.list_cluster_sizes()
```

```python
i = 0
num_curved = zeros(len(cluster_indexes))
num_flat = zeros(len(cluster_indexes))
for cluster in cluster_indexes:
    for curve in cluster:
        if is_curved[curve] == 1:
            num_curved[i] += 1
        else:
            num_flat[i] += 1
    i += 1
```

```python
percent_curved = [num/size for [num, size] in zip(num_curved, cluster_sizes)]
percent_flat = [num/size for [num, size] in zip(num_flat, cluster_sizes)]
```

```python
i = 0
print "Cluster\tSize\t% Flat"
while i < len(cluster_sizes):
    if cluster_sizes[i] > 10 and percent_flat[i] < 0.6 and percent_flat[i] > 0.4:
        print '%d\t%d\t%g' % (F.list_cluster_ids()[i], cluster_sizes[i], round(percent_flat[i],3))
    i = i+1
```
