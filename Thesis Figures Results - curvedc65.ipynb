## Properties of clusters

- take all one peak clusters and find the position of the major peak in unshifted LcD space

```python
print one_peak
print mixed_one
```

```python
one_clusters = []
for id in one_peak:
    one_clusters.append(c.Cluster(F, id))
one_clusters_indexes = [list(clust.list_curve_indexes()) for clust in one_clusters]
one_clusters_indexes.append(mixed_one)
one_lcd_list = [v.cluster_lcd_list(clus_ind, coa) for clus_ind in one_clusters_indexes]
one_peak_locs = []
for lcd_list in one_lcd_list:
    peaks = []
    for lcd in lcd_list:
        maxima_x, maxima_y = findPeaks(lcd[0], lcd[1])
        # fix this in the other one! not max x, x where max y!
        max_x = maxima_x[maxima_y.index(max(maxima_y))]
        peaks.append(max_x)
    one_peak_locs.append(peaks)
all_one_peak_locs = []
for one_peaks in one_peak_locs:
    all_one_peak_locs = all_one_peak_locs + one_peaks
```

```python
print mixed_two
print two_peak
```

```python
two_clusters = []
for id in two_peak:
    two_clusters.append(c.Cluster(F, id))
two_clusters_indexes = [list(clust.list_curve_indexes()) for clust in two_clusters]
two_clusters_indexes.append(mixed_two)
two_lcd_list = [v.cluster_lcd_list(clus_ind, coa) for clus_ind in two_clusters_indexes]
two_peak_locs = []
two_peaks_all_lc = []
two_peaks_all_d = []
for lcd_list in two_lcd_list:
    peaks = []
    all_lc = []
    all_d = []
    for lcd in lcd_list:
        maxima_x, maxima_y = findPeaks(lcd[0], lcd[1], 0.05)
        all_lc.append(maxima_x)
        all_d.append(maxima_y)
        if len(maxima_x) == 1:
            max_x = maxima_x[0]
        else:
            args = argsort(maxima_y)[::-1]
            p1 = args[0]
            p2 = args[1]
            max_x = max(maxima_x[p1], maxima_x[p2])
        peaks.append(max_x)
    two_peak_locs.append(peaks)
    two_peaks_all_lc.append(all_lc)
    two_peaks_all_d.append(all_d)
all_two_peak_locs = []
for two_peaks in two_peak_locs:
    all_two_peak_locs = all_two_peak_locs + two_peaks
```
