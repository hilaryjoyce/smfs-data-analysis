```python
%autoreload 2
cpool = ["#1F78B4", "#E31A1C", "#A6CEE3", "#FB9A99", "#7BCCC4", "#B2DF8A", "#33A02C", "#02818A", "#FF7F00", "#FDBF6F", "#CAB2D6", "#6A3D9A", "#BFD3E6", "#8C96C6"]
import Cluster.clustering as c
```

```python
folder = '../Data/aLa_flat_sample/'
max_x = 200
max_shift = 30
minTss = 10
minForce = 45
p_list = [0.2, 0.36, 0.4, 0.6]
```

```python
param_list = ['%sTransform_analysis/Parameters_p%g_minTss%d_minForce%d/' % (folder, p, minTss, minForce) for p in p_list]
```

```python
coa_list = [c.CoAnalysis(param, max_x, max_shift) for param in param_list]
```

```python
coa0 = coa_list[0]
coa1 = coa_list[1]
coa2 = coa_list[2]
coa3 = coa_list[3]
```

```python
nonzero_list = [ones(len(coa0.co_array))]*4
i = 0
for coa in coa_list:
    k = 0
    for co in coa.co_array:
        if co == 0:
            nonzero_list[i][k] = 0
        k = k+1
    i = i+1
```

```python
coarray_list = [coa.co_array for coa in coa_list]
```

```python
co_nonzero_list = []
i = 0
for co_array in coarray_list:
    nonzero = nonzero_list[i]
    co_nonzero = []
    k = 0
    for z in nonzero:
        if z == 1:
            co_nonzero.append(co_array[k])
        k = k+1
    co_nonzero_list.append(co_nonzero)
    i = i+1
```

```python
hist(co_nonzero_list[0], bins = arange(0,1, 0.02), alpha=0.4);
hist(co_nonzero_list[1], bins = arange(0,1, 0.02), alpha=0.4);
hist(co_nonzero_list[2], bins = arange(0,1, 0.02), alpha=0.4);
hist(co_nonzero_list[3], bins = arange(0,1, 0.02), alpha=0.4);
```

## Comparing p = 0.2 with p = 0.4

```python
co_24_difference = [c4 - c2 for [c2, c4] in zip(co_nonzero_list[0], co_nonzero_list[2])]
```

```python
figsize(6,4)
hist(co_24_difference, bins = arange(-0.3, 0.3, 0.005), normed=True, color=cpool[0], histtype='stepfilled');
title('Coincidence metric differences for\np = 0.2nm and p = 0.4nm', size=14)
text(0.1,16, '$\mu$ = %.1e' % average(co_24_difference), size=14)
text(0.1,14, '$\sigma^2$ = %.1e' % var(co_24_difference), size=14)
xlabel('$\Gamma_{p=0.4}-\Gamma_{p=0.2}$', size=18)
ylabel('Frequency')
print average(co_24_difference)
print std(co_24_difference)
print var(co_24_difference);
```
