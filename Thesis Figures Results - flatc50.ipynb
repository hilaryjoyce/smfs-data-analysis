```python
print one_peak
```

```python
one_clusters = []
for id in one_peak:
    one_clusters.append(c.Cluster(F, id))
one_lcd_list = [v.cluster_lcd_list(clust.list_curve_indexes(), coa) for clust in one_clusters]
one_peak_locs = []
for lcd_list in one_lcd_list:
    peaks = []
    for lcd in lcd_list:
        maxima_x, maxima_y = findPeaks(lcd[0], lcd[1])
        max_x = maxima_x[-1]
        peaks.append(max_x)
    one_peak_locs.append(peaks)
all_one_peak_locs = []
for one_peaks in one_peak_locs:
    all_one_peak_locs = all_one_peak_locs + one_peaks
```
