## Properties of clusters

- major peak locations
- take all one peak clusters and find the position of the major peak in unshifted LcD space

```python
print major_peak
```

```python
major_clusters = []
for id in major_peak:
    major_clusters.append(c.Cluster(F, id))
major_lcd_list = [v.cluster_lcd_list(clust.list_curve_indexes(), coa) for clust in major_clusters]
major_peak_locs = []
for lcd_list in major_lcd_list:
    peaks = []
    for lcd in lcd_list:
        maxima_x, maxima_y = findPeaks(lcd[0], lcd[1])
        # fix this in the other major! not max x, x where max y!
        max_x = maxima_x[maxima_y.index(max(maxima_y))]
        peaks.append(max_x)
    major_peak_locs.append(peaks)
all_major_peak_locs = []
for major_peaks in major_peak_locs:
    all_major_peak_locs = all_major_peak_locs + major_peaks
```

```python
for [id, peak] in zip(major_peak, major_peak_locs):
    print id, peak
```
