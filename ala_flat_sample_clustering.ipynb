## Get flat clusters at co_cut value

```python
co_cut = 0.70
F = c.FlatClusters(coa, co_cut)
print F
```

```python
clusters, sizes = F.largest_clusters(10)
print '%d clusters of %d curves' % (len(clusters), sum(sizes))
print '%.2f%% of data set' % (float(sum(sizes))/coa.get_sample_size()*100)
```

```python
#F.print_clusters_by_size(10)
```

- generate flat clusters and append to fclust list

```python
fclust = []
i = 0
for cluster in clusters:
    fclust.append(c.Cluster(F, cluster))
    print "(%d, %d)" % (cluster, sizes[i]),
    i = i+1
```

## Exploring clusters from flat starting point

```python
figsize(10,14)
i = 1
suptitle("aLa on flat PS: flattened clusters at min coincidence = %g" % F.get_min_coincidence(), size=18);
for flat in fclust:
    print i,
    subplot(9,4,i)
    flat.plot_plain_cluster(alpha=0.5, max_x = 150, max_y=0)
    i = i+1
tight_layout(pad=2);
subplots_adjust(top=0.94)
savefig('%sflat_clusters_c%g_g10.png' % (fig_folder, co_cut))
```

```python
figsize(12,10)
i = 1
suptitle("Flattened clusters at min coincidence = %g" % F.get_min_coincidence(), size=18);
for flat in fclust[0:12]:
    print i,
    subplot(4,3,i)
    flat.plot_cluster(alpha=0.5, max_x = 100)
    i = i+1
tight_layout(pad=2);
subplots_adjust(top=0.95)
savefig('%sflat_clusters_c%g_top12.png' % (fig_folder, co_cut))
```

## Looking at subclusters

```python
i = 4
xp = 80
cluster = fclust[i]
sub = c.SubCluster(F, cluster)
# save a text file listing the curves in this cluster
with open('%scluster%d_list.txt' % (co_folder, cluster.cluster_number), 'w') as file:
    file.write('#index\tname\n')
    i = 0
    cindexes = cluster.list_curve_indexes()
    cnames = cluster.list_curve_names()
    while i < cluster.get_cluster_size():
        file.write('%d\t%s\n' % (cindexes[i], cnames[i]))
        i = i+1
#Plot the whole cluster again
figsize(6,4)
cluster.plot_cluster(alpha=0.3, max_x = xp);
title('Cluster %d' % cluster.cluster_number, size=14);
lcd_y = axis()[3]
savefig('%sLcDensity_cluster%d.png' % (co_folder, cluster.cluster_number))
```
