```python
two_clusters = []
for id in two_peak:
    two_clusters.append(c.Cluster(F, id))
two_clusters_indexes = [list(clust.list_curve_indexes()) for clust in two_clusters]
two_clusters_indexes.append(mixed_two)
two_lcd_list = [v.cluster_lcd_list(clus_ind, coa) for clus_ind in two_clusters_indexes]
two_peak_locs = []
two_peaks_all_lc = []
two_peaks_all_d = []
for lcd_list in two_lcd_list:
    peaks = []
    all_lc = []
    all_d = []
    for lcd in lcd_list:
        maxima_x, maxima_y = findPeaks(lcd[0], lcd[1], 0.2)
        if len(maxima_x) == 1:
            max_x = maxima_x[0]
        else:
            args = argsort(maxima_y)[::-1]
            p1 = args[0]
            p2 = args[1]
            max_x = max(maxima_x[p1], maxima_x[p2])
            all_lc.append(sort([maxima_x[p1], maxima_x[p2]]))
        peaks.append(max_x)
    two_peak_locs.append(peaks)
    two_peaks_all_lc.append(all_lc)
all_two_peak_locs = []
for two_peaks in two_peak_locs:
    all_two_peak_locs = all_two_peak_locs + two_peaks
```
