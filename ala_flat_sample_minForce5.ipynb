## Looking at subclusters

```python
co_folder = coa.shift_folder
print co_folder
```

```python
F = c.FlatClusters(coa, co_cut = 0.75)
clusters, sizes = F.largest_clusters(10)
fclust = []
i = 0
for cluster in clusters:
    fclust.append(c.Cluster(F, cluster))
    print "(%d, %d)" % (cluster, sizes[i]),
    i = i+1
print len(fclust)
```

```python
j = 0
while j < len(fclust):
    xp = 150
    cluster = fclust[j]
    # save a text file listing the curves in this cluster
    cindexes = cluster.list_curve_indexes()
    cnames = cluster.list_curve_names()
    with open('%scluster%d_list.txt' % (co_folder, cluster.cluster_number), 'w') as file:
        file.write('#index\tname\n')
        i = 0
        while i < cluster.get_cluster_size():
            file.write('%d\t%s\n' % (cindexes[i], cnames[i]))
            i = i+1
    #Plot the whole cluster again
    figsize(6,4)
    cluster.plot_cluster(alpha=0.3, max_x = xp);
    title('Cluster %d' % cluster.cluster_number, size=14);
    lcd_y = axis()[3]
    savefig('%sLcDensity_cluster%d.png' % (co_folder, cluster.cluster_number))
    clf()
    print j,
    j = j+1
```

```python
print k
cluster = fclust[k]
tf_y = 500
max_x = 100
lcd_y = 0.2
lcd_list = cluster.get_Lc_density_arrays()
tf_list = cluster.get_tss_force_arrays()
num_list = cluster.list_curve_indexes()
cols = 4
rows = ceil(len(lcd_list)/4.0)
height = rows * 2.2
width = 12
#print len(lcd_list), rows, height
#fig = figure(k)
figsize(width, height)
i = 1
while i <=len(lcd_list):
    lcd = lcd_list[i-1];
    tf = tf_list[i-1];
    lc_ax = subplot(rows, cols, i);
    lc_ax.plot(lcd[0], lcd[1], color='k', lw=1.5);
    lc_ax.set_ylim(0, lcd_y);
    lc_ax.text(max_x/1.2, lcd_y/1.2, '%d' % num_list[i-1])
    tf_ax = lc_ax.twinx()
    tf_ax.plot(tf[0], tf[1], color='k', alpha=0.4)
    tf_ax.set_ylim(-100, tf_y)
    lc_ax.set_xlim(0, max_x);
    tf_ax.set_xticks([])
    tf_ax.set_yticks([])
    lc_ax.set_xticks([])
    lc_ax.set_yticks([])
    i = i+1;
tight_layout()
savefig('%sCluster%d_subplots.png' % (co_folder, cluster.cluster_number))
k = k+1
```
