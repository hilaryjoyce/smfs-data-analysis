```python
def absCoincidence(d1, d2):
    '''Runs in one function. Most efficient.'''
    from scipy.signal import fftconvolve
    from numpy import dot, sqrt, where, log10

    fft_flip = fftconvolve(d1, d2[::-1], mode='full')
    max_conv = max(fft_flip)
    max_id = where(fft_flip == max_conv)[0][0]     
    max_id = max_id - len(fft_flip)/2
    dot_d1 = dot(d1, d1)
    dot_d2 = dot(d2, d2)

    if (dot_d1 <= dot_d2):
        s = sqrt(dot_d1/dot_d2)
    else:
        s = sqrt(dot_d2/dot_d1)
    
    print '\ts =', s
    print '\tUnscaled gamma =', max_conv / sqrt(dot_d1 * dot_d2)
    Gamma = max_conv / sqrt(dot_d1 * dot_d2) * s
    print '\tGamma =', Gamma
    return round(Gamma,5), max_id

def CoincidenceList(force_list):
    '''
    Calculates a list of the maximum coincidence and corresponding shift 
    for ONE particular shift.
    '''
    from glob import glob
    densityFiles = glob("%sDensity_text/*.txt" % folder)
    
    eps = 10**(-3) # fixinig this since I haven't changed it in so long I forget what it means...

    N = len(densityFiles)
    count = 0
    i = 0
    curve1_list = []
    curve2_list = []
    gamma_list = []
    umax_list = []
    while (i < len(force_list)):
        d1 = force_list[i]
        curveNum1 = '%d' % i
        print curveNum1,
        k = i + 1
        while (k < len(force_list)):
            curveNum2 = '%d' % k
            print curveNum2
            d2 = force_list[k]
            if (max(d1) == 0 or max(d2) == 0):
                if (max(d1) == 0 and max(d2) == 0):
                    Gamma, u_max, x_max = 1.0, 0, 0
                else:
                    Gamma, u_max, x_max = 0, 0, 0
            else:
                Gamma, u_max = absCoincidence(d1, d2)
            curve1_list.append(curveNum1)
            curve2_list.append(curveNum2)
            gamma_list.append(Gamma)
            umax_list.append(u_max)
            k = k+1
        i += 1

    return curve1_list, curve2_list, gamma_list, umax_list

def plot_dendrogram(Z):
    '''
    Plots the dendragram visualization of Z and returns
    the dendrogram object 'dendro'.
    '''
    from scipy.cluster.hierarchy import dendrogram, set_link_color_palette
    import matplotlib.pyplot as plt

    cpool = ["#1F78B4", "#E31A1C", "#A6CEE3", "#FB9A99", "#7BCCC4", "#B2DF8A", "#33A02C", "#02818A", "#FF7F00", "#FDBF6F", "#CAB2D6", "#6A3D9A", "#BFD3E6", "#8C96C6"]
    set_link_color_palette(cpool)

    dendro = dendrogram(Z, no_labels=False, 
        count_sort=True, orientation="left");
    plt.xlabel("Coincidence Metric ($\Gamma$)", fontsize = 14)
    plt.ylabel("Clusters", fontsize = 14)
    plt.xticks([1, 0.8, 0.6, 0.4, 0.2, 0], [0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.xticks(fontsize=14)
    return dendro
```

```python
# generic tss_array
master_tss = arange(0, max_tss, 0.1)
```

```python
# function to take some tss value, a tss and force array, and return the average force at that tss value?
def av_force(tss, force, location, d = 0.5):
    region = where(all(zip(tss < location+d, tss > location-d), axis=1))[0]
    if len(region)>0:
        return average(force[region])
    else:
        return 0

def running_average(tss, force, master_tss, d = 0.5):
    average_force = zeros(len(master_tss))
    i = 0
    while master_tss[i] < tss[0]:
        i = i+1
    while master_tss[i] < tss[-1]:
        average_force[i] = av_force(tss, force, master_tss[i], d)
        i = i+1
    return average_force
```
