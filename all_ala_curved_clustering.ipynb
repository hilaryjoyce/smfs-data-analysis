## Working area for getting average cluster density

```python
lc_d= s33_4.get_Lc_density_arrays('left')
ishifts = s33_4.list_initial_shifts('left')
density = [x[1] for x in lc_d]
```

```python
def cluster_Lc_shifted(ishifts, Lc_density_list):
    shift_list = ishifts
    Lc_list = [x[0] for x in Lc_density_list]
    initial_shifts = ishifts
    return [[Lc+shift for Lc in Lc_array] for [Lc_array, shift] in zip(Lc_list, initial_shifts)]
```

```python
def shift_to_index(shift, grid=0.25):
    return int(shift/grid)
```

```python
shift_Lc = cluster_Lc_shifted(ishifts, lc_d)
```

```python
def average_cluster_density(ishifts, Lc_density_list):
    ''' Takes a cluster number and a set of flat clusters, the shift details and the 
    density list, and returns the average_density '''
    from numpy import copy
    shift_list = ishifts
    density_list = [x[1] for x in Lc_density_list]
    N = len(density_list)
    M = len(density_list[0])
    average_density = copy(density_list[0])
    k = 1
    while k < N:
        ind_shift = shift_to_index(shift_list[k-1])
        if ind_shift <= 0:
            j = 0
            while j < M+ind_shift:
                average_density[j] = average_density[j] + density_list[k][j-ind_shift]
                j = j+1
        else:
            j = ind_shift
            while j < M:
                average_density[j] = average_density[j] + density_list[k][j-ind_shift]
                j = j+1
        k = k+1
    average_density = average_density/N
    return average_density
```

```python
av_d = average_cluster_density(ishifts, lc_d)
```

```python
figsize(6,4)
i = 0
while i < len(ishifts):
    plot(shift_Lc[i], density[i], 'k-', alpha=0.5)
    i = i+1
plot(shift_Lc[0], av_d, 'r-', lw=2)
xlim(0,60)
```
